```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
X = pd.read_csv(".\\data\\home-data-for-ml-course\\train.csv", index_col='Id')
X.head()
```

```python
X_test = pd.read_csv(".\\data\\home-data-for-ml-course\\test.csv", index_col='Id')
X_test.head()
```

```python
# Remove missing value row 

X.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X.SalePrice
X.drop(['SalePrice'], axis=1, inplace=True)
```

```python
cols_with_missing = [col for col in X.columns if X[col].isnull().any()]
cols_with_missing
```

```python
X.drop(cols_with_missing, axis=1, inplace=True)
X_test.drop(cols_with_missing, axis=1, inplace=True)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_model(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(random_state=0, n_estimators=100)    
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, preds)
    return mae
```

```python
s = X_train.select_dtypes(include=['object']).columns
s
```

```python
s = (X_train.dtypes == 'object')
s = s[s]
s

drop_X_train = X_train.copy()
drop_X_valid = X_valid.copy()
```

```python
drop_X_train.drop(s.index, axis=1, inplace=True)  # drop object columns from X_train
drop_X_valid.drop(s.index, axis=1, inplace=True)  # drop object columns from X_valid
drop_X_train.columns
```

```python
print("MAE dropped Categorical: " ,score_model(drop_X_train, drop_X_valid, y_train, y_valid))
```

```python
print("Unique values in 'Condition2' column in training data:", X_train['Condition2'].unique())
print("Unique values in 'Condition2' column in validation data:", X_valid['Condition2'].unique())
```

```python
# Select all categorical columns
obj_col_X_train = [col for col in X_train.columns if X_train[col].dtype == 'object']
obj_col_X_train

# Columns that can be safely label encoded
good_obj_col_X_train = [col for col in obj_col_X_train if set(X_train[col].unique()) == set(X_valid[col].unique())]
good_obj_col_X_train
```

```python
bad_obj_col_X_train = list(set(obj_col_X_train) - set(good_obj_col_X_train))
bad_obj_col_X_train
```

```python

print('Categorical columns that will be label encoded:', good_obj_col_X_train)
print('\nCategorical columns that will be dropped from the dataset:', bad_obj_col_X_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_X_train = X_train.drop(bad_obj_col_X_train, axis=1).copy()
label_X_valid = X_valid.drop(bad_obj_col_X_train, axis=1).copy()
label_X_train.shape, label_X_valid.shape
```

```python
label_encoder = LabelEncoder()

for col in good_obj_col_X_train:
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    label_X_valid[col] = label_encoder.transform(X_valid[col])
    
```

```python
print("MAE with label encoding: ", score_model(label_X_train, label_X_valid, y_train, y_valid))
```
